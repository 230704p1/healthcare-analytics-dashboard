<h1>HHS care Trend Graph</h1>

```python
df['Date'] = pd.to_datetime(df['Date'])

monthly_hhs = df.groupby(
    df['Date'].dt.to_period('M')
)['Children in HHS Care'].mean()
```

```python
monthly_hhs.index = monthly_hhs.index.astype(str)

plt.figure(figsize=(14,6))

plt.plot(
    monthly_hhs.index,
    monthly_hhs.values,
    marker='o',
    linewidth=3
)

plt.title("Monthly Average of Children in HHS Care")
plt.xlabel("Month")
plt.ylabel("Average Children in HHS Care")

plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

<h1>Children Apprehended</h1>

```python
plt.figure(figsize=(12,5))

plt.bar(df['Date'].head(20),
        df['Children apprehended and placed in CBP custody*'].head(20))

plt.title("Children Apprehended")
plt.xticks(rotation=90)

plt.show()
```

<h1>CBP Custody vs HHS Care</h1>

```python


plt.figure(figsize=(10,6))

# Scatter Plot
plt.scatter(
    df['Children in CBP custody'],
    df['Children in HHS Care'],
    color='skyblue',
    alpha=0.6,
    edgecolors='black'
)

# Trend Line
z = np.polyfit(
    df['Children in CBP custody'],
    df['Children in HHS Care'],
    1
)

p = np.poly1d(z)

plt.plot(
    df['Children in CBP custody'],
    p(df['Children in CBP custody']),
    color='red',
    linewidth=3
)

plt.xlabel(
    "Children in CBP Custody",
    fontsize=12,
    fontweight='bold'
)

plt.ylabel(
    "Children in HHS Care",
    fontsize=12,
    fontweight='bold'
)

plt.title(
    "CBP Custody vs HHS Care",
    fontsize=20,
    fontweight='bold'
)

plt.grid(True)

plt.show()
```

<h1>Distribution</h1>

```python
plt.figure(figsize=(10,6))

plt.hist(
    df['Children in HHS Care'],
    bins=15,
    color='purple',
    edgecolor='black',
    alpha=0.7
)

plt.title(
    "Distribution of HHS Care",
    fontsize=20,
    fontweight='bold'
)

plt.xlabel(
    "Children in HHS Care",
    fontsize=12,
    fontweight='bold'
)

plt.ylabel(
    "Frequency",
    fontsize=12,
    fontweight='bold'
)

# Remove crowded x-ticks
plt.xticks([])

plt.grid(True)

plt.show()
```

```python
df['Children in CBP custody'] = pd.to_numeric(
    df['Children in CBP custody'],
    errors='coerce'
)

df['Children in HHS Care'] = pd.to_numeric(
    df['Children in HHS Care'],
    errors='coerce'
)
```

<h1>Total System Load Chart</h1>

```python


# Group by Month
monthly_load = df.groupby(
    df['Date'].dt.to_period('M')
)['Total System Load'].mean()

# Convert Period to String
monthly_load.index = monthly_load.index.astype(str)

# Plot Graph
plt.figure(figsize=(14,6))

plt.plot(
    monthly_load.index,
    monthly_load.values,
    color='red',
    marker='o',
    linewidth=3
)

plt.title(
    "Monthly Average Total System Load",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel("Month")
plt.ylabel("Average Total Children")

plt.grid(True, linestyle='--', alpha=0.7)

# Show only some labels
plt.xticks(
    range(0, len(monthly_load.index), 3),
    monthly_load.index[::3],
    rotation=45
)

plt.tight_layout()
plt.show()
```

<h1>Children Discharged from HHS Care Trend</h1>

```python
monthly_discharge = df.groupby(
    df['Date'].dt.to_period('M')
)['Children discharged from HHS Care'].mean()

monthly_discharge.index = monthly_discharge.index.astype(str)

plt.figure(figsize=(14,6))

plt.plot(
    monthly_discharge.index,
    monthly_discharge.values,
    marker='o',
    linewidth=3
)

plt.title("Monthly Average Children Discharged from HHS Care")

plt.xlabel("Month")
plt.ylabel("Average Discharged Children")

plt.xticks(
    range(0, len(monthly_discharge.index), 3),
    monthly_discharge.index[::3],
    rotation=45
)

plt.grid(True)

plt.tight_layout()
plt.show()
```

<h1>CBP Custody Trend Over Time</h1>

```python
monthly_cbp = df.groupby(
    df['Date'].dt.to_period('M')
)['Children in CBP custody'].mean()

monthly_cbp.index = monthly_cbp.index.astype(str)

plt.figure(figsize=(14,6))

plt.plot(
    monthly_cbp.index,
    monthly_cbp.values,
    color='orange',
    marker='o',
    linewidth=3
)

plt.title("Monthly Average Children in CBP Custody")

plt.xlabel("Month")
plt.ylabel("Average Children")

plt.xticks(
    range(0, len(monthly_cbp.index), 3),
    monthly_cbp.index[::3],
    rotation=45
)

plt.grid(True)

plt.tight_layout()
plt.show()
```
